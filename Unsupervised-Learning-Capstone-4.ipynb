```python
%matplotlib inline
import numpy as np
import pandas as pd
import scipy
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
filename=['And Thus He Came.txt',
         'Christmas Day in the Morning.txt']
```

```python
def text_creater(filename):
    file=open(filename,'r')
    text=file.read()
    file.close
    return text
```

```python
text=text_creater(filename[0])
```

## Text Cleaning

```python
def text_cleaner(text):
    text=re.sub(r'ï»¿','',text)
    text=re.sub(r'\n','',text)
    text=re.sub('\[[^()]*\]','',text)
    text=re.sub('\([^()]*\)','',text)
    text=re.sub('[^\w\.\,]',' ',text)
    text=' '.join(text.split())
    return text
```

```python
brady=text_cleaner(text)
```

```python
import spacy
nlp = spacy.load('en')
```

```python
# All the processing work is done here, so it may take a while.
text_doc = nlp(brady)
```

```python
text_sents=[[sent,"brady"] for sent in text_doc.sents]
text_sents
```

```python
# Combine the sentences from the two novels into one data frame.
sentences = pd.DataFrame(text_sents)
```
